```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[49,50,53,44,46,48,45]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'b+--')
```

```python
plt.plot(x,y,color='red',marker='+',linestyle='')
```

```python
plt.plot(x,y,color='orange',marker='*',linestyle='',markersize=10)
```

```python
plt.plot(x,y,'g<',alpha=0.6)
```
